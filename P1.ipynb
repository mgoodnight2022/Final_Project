```python
import requests
import pandas as pd
import json
```

```python
res = requests.get("https://api.open-meteo.com/v1/forecast?latitude=31.48&longitude=-83.53&hourly=temperature_2m,precipitation,windspeed_10m,winddirection_10m,soil_moisture_0_1cm,soil_moisture_1_3cm,soil_moisture_3_9cm,soil_moisture_9_27cm,soil_moisture_27_81cm&daily=sunrise,sunset,precipitation_sum,precipitation_hours,windspeed_10m_max,winddirection_10m_dominant&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FNew_York&past_days=2")
```

```python
print(res.json)
print(list(res.json()))
```

```python
hourlyunits_json = res.json()
#print(hourlyunits_json)
hourlyunits_json["hourly_units"]
df_hourly = pd.DataFrame(hourlyunits_json["hourly_units"], index = ["1"])
display(df_hourly)
```

```python
#Has it rained?
hourly_json = res.json()
hourly_json["hourly"]
df_hourly = pd.DataFrame(hourly_json["hourly"])
display(df_hourly)

df_rain = df_hourly.sort_values(by = ["time"], ascending = True)
#display(df_rain)

df_precip = df_rain[(df_rain['precipitation']>0)]

print("Here is the hourly precipitation in inches for the last week. No need to run the irrigation during rain events!")
display(df_precip)
```

```python
daily_json = res.json()
daily_json["daily"]
df_daily = pd.DataFrame(daily_json["daily"])
display(df_daily)
```
