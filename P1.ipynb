```python
import requests
import pandas as pd
import json
```

```python
res = requests.get("https://api.open-meteo.com/v1/forecast?latitude=40.71&longitude=-74.01&hourly=temperature_2m,precipitation&daily=sunrise,sunset&temperature_unit=fahrenheit&windspeed_unit=mph&precipitation_unit=inch&timezone=America%2FNew_York&past_days=2")
```

```python
print(list(res.json()))
```

```python
hourly_json = res.json()
hourly_json["hourly"]
df_hourly = pd.DataFrame(hourly_json["hourly"])
print(df_hourly)
```

```python
daily_json = res.json()
daily_json["daily"]
df_daily = pd.DataFrame(daily_json["daily"])
print(df_daily)
```
